```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
import pandas as pd
import numpy as np


train = pd.read_csv('train.csv',index_col='ticket_id',encoding = 'ISO-8859-1' )
test = pd.read_csv('test.csv',index_col='ticket_id',encoding='ISO-8859-1')
train.dropna(subset=['compliance'],how='any',inplace=True)
def row(x):
    x['ticket_issued_date'] = x['ticket_issued_date'].split(' ')[0]
    x['hearing_date'] = x['hearing_date'].split(' ')[0]
    return x
train['hearing_date'].fillna(method = 'ffill',inplace=True) ##forward fill the NaN in hearing_date
test['hearing_date'].fillna(method = 'ffill',inplace=True)
train = train.apply(row,axis=1)  ##delete the detialed hour
test = test.apply(row,axis=1)

train['issue_hearing_duration(days)'] = (pd.to_datetime(train['hearing_date'])-
                                   pd.to_datetime(train['ticket_issued_date'])).dt.days  #calculate the duration
test['issue_hearing_duration(days)'] = (pd.to_datetime(test['hearing_date'])-
                                   pd.to_datetime(test['ticket_issued_date'])).dt.days  #calculate the duration

dummy_variables = pd.get_dummies(train[['agency_name','state']])
dummy_variables_test = pd.get_dummies(test[['agency_name','state']])

dummy_variables,dummy_variables_test = dummy_variables.align(dummy_variables_test, join='outer', axis=1, fill_value=0)

address = pd.read_csv('addresses.csv',encoding='ISO-8859-1')
latlons = pd.read_csv('latlons.csv',encoding='ISO-8859-1')
address_latlons = pd.merge(left=address,right=latlons,how='outer',left_on='address',right_on='address')

#merge the old train with latitude and lontitude information
new_train = pd.merge(left=train,right=address_latlons,how='left',left_index=True,right_on='ticket_id')
new_test = pd.merge(left=test,right=address_latlons,how='left',left_index=True,right_on='ticket_id')
#construnct the continuous variable

y_train = new_train['compliance']
new_train = new_train[['ticket_id','fine_amount','admin_fee','state_fee','late_fee',
                       'discount_amount','clean_up_cost','judgment_amount',
                       'issue_hearing_duration(days)','lat','lon']]
new_test = new_test[['ticket_id','fine_amount','admin_fee','state_fee','late_fee',
                       'discount_amount','clean_up_cost','judgment_amount',
                       'issue_hearing_duration(days)','lat','lon']]
new_train.fillna(method='ffill',inplace=True)
new_test.fillna(method='ffill',inplace=True)

#normalization
scaler = MinMaxScaler()
X_train_continous = scaler.fit_transform(new_train.iloc[:,1:])
X_train_continous_labeled = pd.DataFrame(data=X_train_continous,
                                         index=new_train['ticket_id'],
                                         columns=new_train.columns[1:])

X_test_continous = scaler.transform(new_test.iloc[:,1:])
X_test_continous_labeled = pd.DataFrame(data=X_test_continous,
                                        index=new_test['ticket_id'],
                                        columns=new_test.columns[1:])

X_train = pd.merge(left=X_train_continous_labeled, right=dummy_variables, 
                   how='left', left_index=True,right_index=True).iloc[:,1:]
X_test = pd.merge(left=X_test_continous_labeled, right=dummy_variables_test, 
                  how='left', left_index=True,right_index=True).iloc[:,1:]
```
